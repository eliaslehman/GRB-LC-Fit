```python
### BPL Break Flux ###
def Fa(F_p, alpha_p, alpha_a1, t0, t1):
    return F_p - alpha_p * t0 - alpha_a1 * (t1 - t0)

### Error Propagation for BPL Break Flux ###
def dFa(F_p, alpha_p, alpha_a1, t0, t1, dF_p, dalpha_p, dalpha_a1, dt0, dt1):
    return np.sqrt(dF_p**2+t0**2*dalpha_p**2+(-alpha_p+alpha_a1)**2*dt0**2+(t0-t1)**2*dalpha_a1**2+alpha_a1**2*dt1**2)
```
